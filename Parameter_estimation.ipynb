```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
from sociophysicsDataHandler import SociophysicsDataHandler
import os
from tqdm import tqdm
from Functions.Hamiltonian_Solver import *
from Functions.Parameter_Estimator import *
from scipy.signal import savgol_filter
from scipy.optimize import curve_fit
from mpl_toolkits.mplot3d import Axes3D

%load_ext autoreload
%autoreload 2
```

```python
os.chdir("C:/Users")
os.chdir("20183572")
os.chdir("SURFdrive/Shared/trajs-by-door-id")
files = os.listdir()
```

```python
dx = 0.1
v_tresh = 0.1
max_x = 6
max_y = 2
bound = [-max_x*1000,max_x*1000,-max_y*1000,100]

matrix = np.zeros([len(files), int(2*max_x/dx), int(2*max_y/dx)])

def y_vel(df):
    window_length = 11
    poly_order = 1
    df = df.assign(**{"traj_len": df.groupby('tracked_object')['tracked_object'].transform("size")})
    df = df[df["traj_len"] > window_length].copy()
    df["y_vel"] = df.groupby('tracked_object').y_pos.transform(lambda x: savgol_filter(x, window_length=window_length, polyorder=poly_order, deriv=1))
    return df

for i in tqdm(range(len(files))):
    filt_data = pd.read_csv(files[i]).copy()
    bm_bound = (filt_data["x_pos"] > bound[0]) & (filt_data["x_pos"] < bound[1]) & (filt_data["y_pos"] > bound[2]) & (filt_data["y_pos"] < bound[3])
    filt_data = filt_data[bm_bound].reset_index(drop = True)
    filt_data = y_vel(filt_data)
    bm_vel = (filt_data["y_vel"] > -v_tresh) & (filt_data["y_vel"] < v_tresh)
    filt_data = filt_data[bm_vel].reset_index(drop = True)

    filt_data["x_pos"] = (filt_data["x_pos"] /1000) //dx
    filt_data["y_pos"] = (filt_data["y_pos"] /1000) //dx
    
    static_sum = filt_data.groupby(by = ["timestampms"])["tracked_object"].nunique()
    filt_data = filt_data.join(static_sum, on='timestampms', rsuffix='_sum')

    bm_highest = filt_data["tracked_object_sum"] == filt_data["tracked_object_sum"].max()

    static = filt_data[bm_highest].groupby(by = ["x_pos", "y_pos"])["tracked_object"].nunique()
    
    if static.count() > 1: 
        xy = np.array(list(map(np.array,static.index.values)),dtype=np.int32)
        N = static.values

        matrix[i, xy[:,0] - int(max_x/dx), xy[:,1] - int(max_y/dx)] = N
```

```python
g_exp = matrix.mean(axis=0)
```

```python
%matplotlib inline

fig = plt.figure(figsize = (2*max_x,2*max_y))
ax = plt.axes()
m = ax.imshow(g_exp.T/(dx**2), cmap = "Greys")
ax.invert_yaxis()
cb=fig.colorbar(m)
cb.ax.get_yaxis().labelpad = 30
cb.ax.set_ylabel(r"Density $\rho_\ell$", size = 18, rotation = 270)
ax.set_xlabel(r"$\ell_x$", size = 20)
ax.set_ylabel(r"$\ell_y$", size = 20)
ax.axhline(22,c="k")
ax.annotate("Door", [55,23], size = 12)
ax.set_ylim(0,26)
fig.savefig("Platform_distribution.png")
```

```python
n = 10
mu = g_exp.max()
z = 2
t = 0.1
dim = np.array([g_exp.shape[0], g_exp.shape[1]])
def U_func(x):
    return 1
def V_func(x, *p):
    d = np.divide(x, dim)
    d1 = d - 1/2 + np.array([-p[0],p[1]])
    d2 = d - 1/2 + np.array([p[0],p[1]])
    return mu*(1-(np.exp(-np.sum((d1/p[2:4])**2, axis = 1)) + np.exp(-np.sum((d2/p[2:4])**2, axis = 1))))
def g_func(x, *p):
    return (mu - V_func(x, *p))/U_func(x)
```

```python
X,Y = np.meshgrid(linspace(1,g_exp.shape[0],g_exp.shape[0]),linspace(1,g_exp.shape[1],g_exp.shape[1]))
X = X.T
Y = Y.T
xdata = np.vstack([X.ravel(), Y.ravel()]).T
```

```python
params, pcov = curve_fit(g_func,xdata,g_exp.ravel(),[0.1,0.2, 0.2, 0.2], maxfev = 10000)
sigmas = np.sqrt(np.diag(pcov))
```

```python
params, sigmas
```

```python
g_fit =  g_func(xdata,*params).reshape(g_exp.shape[0],g_exp.shape[1])

fig = plt.figure(figsize = (9,6))
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, g_fit/dx**2, cmap='plasma')
ax.set_xlabel(r"$\ell_x$", size = 18)
ax.set_ylabel(r"$\ell_y$", size = 18)
ax.set_zlabel(r"$\rho_\ell$", size = 18)
fig.savefig("Density_fit.png")
```
